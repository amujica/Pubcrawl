## Terrorist Network Model - One-at-time approach

```python
input_data = pickle.load(open("TerroristNetworkModel/pickles/One_at_time_Network.pickle", "rb"))
spread = []
radicalism = []
for i in range(len([name for name in os.listdir("TerroristNetworkModel/oat_sim/") if os.path.isfile(os.path.join("TerroristNetworkModel/oat_sim/", name))])):
    G = pickle.load(open("TerroristNetworkModel/oat_sim/One_at_time_Network_sim_{}.pickle".format(i), "rb"))
    initial_infected = 0
    final_infected = 0
    mean_belief = 0
    for node in range(G.number_of_nodes()):
        for id in G.nodes[node]['attr_id']:
            if id[0] == 1 and id[1] == 0 and id[2] != 0: initial_infected += 1
            if id[0] == 2 and id[1] == 0 and id[2] != 0: initial_infected += 1
            if id[0] == 1 and id[2] == 200: final_infected += 1 # Terrorists
            if id[0] == 2 and id[2] == 200: final_infected += 1 # Leaders
        for id in G.nodes[node]['attr_mean_belief']:
            if id[2] == 200: mean_belief += id[0]
    
    spread.append((final_infected - initial_infected) / (G.number_of_nodes() - initial_infected))
    radicalism.append(mean_belief / G.number_of_nodes())
```

## Terrorist Network Model - Morris Problem

Using SALib to run the Morris Problem

### Import the package

```python
from SALib.sample import morris as ms
from SALib.analyze import morris as ma
from SALib.plotting import morris as mp
```

```python
spread = []
radicalism = []
for i in range(len([name for name in os.listdir("TerroristNetworkModel/Morris_sim/") if os.path.isfile(os.path.join("TerroristNetworkModel/Morris_sim/", name))])):
    G = pickle.load(open("TerroristNetworkModel/Morris_sim/Morris_Network_sim_{}.pickle".format(i), "rb"))
    initial_infected = 0
    final_infected = 0
    mean_belief = 0
    for node in range(G.number_of_nodes()):
        for id in G.nodes[node]['attr_id']:
            if id[0] == 1 and id[1] == 0 and id[2] != 0: initial_infected += 1
            if id[0] == 2 and id[1] == 0 and id[2] != 0: initial_infected += 1
            if id[0] == 1 and id[2] == 200: final_infected += 1 # Terrorists
            if id[0] == 2 and id[2] == 200: final_infected += 1 # Leaders
        for id in G.nodes[node]['attr_mean_belief']:
            if id[2] == 200: mean_belief += id[0]
    
    spread.append((final_infected - initial_infected) / (G.number_of_nodes() - initial_infected))
    radicalism.append(mean_belief / G.number_of_nodes())
    
data = pickle.load(open("TerroristNetworkModel/pickles/Morris_Network.pickle", "rb"))
morris_problem = data['morris_problem']
sample = data['sample']
```

### Define a problem

In the code below, a problem file is used to define the variables we wish to explore

```python
morris_problem = {
    # There are seven variables
    'num_vars': 8,
    # These are their names
    'names': [  'information_spread_intensity',
                'terrorist_additional_influence',
                'max_vulnerability',
                'prob_interaction',
                'vision_range',
                'sphere_influence',
                'weight_social_distance',
                'weight_link_distance'
             ],
    # Plausible ranges over which we'll move the variables
    'bounds': [[0, 1],  # information_spread_intensity
               [0, 1],  # terrorist_additional_influence
               [0, 1],  # max_vulnerability
               [0, 1],  # prob_interaction
               [0, 1],  # vision_range
               [1, 10], # sphere_influence
               [0, 1],  # weight_social_distance
               [0, 1]   # weight_link_distance
              ],
    # I don't want to group any of these variables together
    'groups': None
    }
```

### Generate a Sample

We then generate a sample using the morris.sample() procedure from the SALib package.

```python
number_of_trajectories = 200
sample = ms.sample(morris_problem, number_of_trajectories, num_levels=4, grid_jump=2)

# pickle.dump({
#     'morris_problem': morris_problem,
#     'sample': sample
# }, open( "pickles/Morris_Network.pickle", "wb" ))
```

```python
# Run the code
monte_carlo_large(sample.T)
```

### Factor Prioritisation

We'll run a sensitivity analysis of the power module to see which is the most influential parameter.

The results parameters are called **mu**, **sigma** and **mu_star**.

- **Mu** is the mean effect caused by the input parameter being moved over its range.
- **Sigma** is the standard deviation of the mean effect.
- **Mu_star** is the mean absolute effect.

```python
Si = ma.analyze(morris_problem, sample, np.array(spread), print_to_console=False)
df = pd.DataFrame(data=list(zip(morris_problem['names'], Si['mu'], Si['mu_star'], Si['sigma'])), 
                  columns=['Name', 'mu', 'mu_star', 'sigma'])
df.style.set_table_styles([
    {'selector': '.row_heading, .blank', 'props': [('display', 'none;')]}
])
```

We can plot the results.

```python
plt.rcParams["figure.figsize"] = (30,15)
fig, (ax1, ax2) = plt.subplots(1,2)
mp.horizontal_bar_plot(ax1, Si, {})
mp.covariance_plot(ax2, Si, {})
```

## Scale Free - Morris Problem

```python
spread = []
radicalism = []
for i in range(len([name for name in os.listdir("NetworkTopology_Size/ScaleFree_sim/") if os.path.isfile(os.path.join("NetworkTopology_Size/ScaleFree_sim/", name))])):
    G = pickle.load(open("NetworkTopology_Size/ScaleFree_sim/ScaleFree_sim_{}.pickle".format(i), "rb"))
    initial_infected = 0
    final_infected = 0
    mean_belief = 0
    for node in range(G.number_of_nodes()):
        for id in G.nodes[node]['attr_id']:
            if id[0] == 1 and id[1] == 0 and id[2] != 0: initial_infected += 1
            if id[0] == 2 and id[1] == 0 and id[2] != 0: initial_infected += 1
            if id[0] == 1 and id[2] == 200: final_infected += 1 # Terrorists
            if id[0] == 2 and id[2] == 200: final_infected += 1 # Leaders
        for id in G.nodes[node]['attr_mean_belief']:
            if id[2] == 200: mean_belief += id[0]
    
    try:
        spread.append((final_infected - initial_infected) / (G.number_of_nodes() - initial_infected))
    except ZeroDivisionError:
        spread.append(0)
    radicalism.append(mean_belief / G.number_of_nodes())
    
data = pickle.load(open("NetworkTopology_Size/pickles/ScaleFree.pickle", "rb"))
morris_problem = data['morris_problem']
sample = data['sample']
```

```python
morris_problem = {
    # There are seven variables
    'num_vars': 2,
    # These are their names
    'names': [  'nodes',
                'civilian_weight'
             ],
    # Plausible ranges over which we'll move the variables
    'bounds': [[50, 500], # Nodes
               [0, 1]      # Civilian Weight
              ],
    # I don't want to group any of these variables together
    'groups': None
    }

number_of_trajectories = 100
sample = ms.sample(morris_problem, number_of_trajectories, num_levels=4, grid_jump=2)
```

```python
# Run the code
monte_carlo_large(sample.T)
```

```python
Si = ma.analyze(morris_problem, sample, np.array(spread), print_to_console=False)
df = pd.DataFrame(data=list(zip(morris_problem['names'], Si['mu'], Si['mu_star'], Si['sigma'])), 
                  columns=['Name', 'mu', 'mu_star', 'sigma'])
df.style.set_table_styles([
    {'selector': '.row_heading, .blank', 'props': [('display', 'none;')]}
])
```

```python
plt.rcParams["figure.figsize"] = (30,15)
fig, (ax1, ax2) = plt.subplots(1,2)
mp.horizontal_bar_plot(ax1, Si, {})
mp.covariance_plot(ax2, Si, {})
```

## Small World - Morris Problem

### Mean radicalism

It is computed as the average radicalism of the agents within the network.

```python
spread = []
radicalism = []
for i in range(len([name for name in os.listdir("NetworkTopology_Size/SmallWorld_sim/") if os.path.isfile(os.path.join("NetworkTopology_Size/SmallWorld_sim/", name))])):
    G = pickle.load(open("NetworkTopology_Size/SmallWorld_sim/SmallWorld_sim_{}.pickle".format(i), "rb"))
    initial_infected = 0
    final_infected = 0
    mean_belief = 0
    for node in range(G.number_of_nodes()):
        for id in G.nodes[node]['attr_id']:
            if id[0] == 1 and id[1] == 0 and id[2] != 0: initial_infected += 1
            if id[0] == 2 and id[1] == 0 and id[2] != 0: initial_infected += 1
            if id[0] == 1 and id[2] == 200: final_infected += 1 # Terrorists
            if id[0] == 2 and id[2] == 200: final_infected += 1 # Leaders
        for id in G.nodes[node]['attr_mean_belief']:
            if id[2] == 200: mean_belief += id[0]
    
    try:
        spread.append((final_infected - initial_infected) / (G.number_of_nodes() - initial_infected))
    except ZeroDivisionError:
        spread.append(0)
    radicalism.append(mean_belief / G.number_of_nodes())
    
data = pickle.load(open("NetworkTopology_Size/pickles/SmallWorld.pickle", "rb"))
morris_problem = data['morris_problem']
sample = data['sample']
```

```python
morris_problem = {
    # There are seven variables
    'num_vars': 4,
    # These are their names
    'names': [  'nodes',
                'joints',
                'probability',
                'civilian_weight'
             ],
    # Plausible ranges over which we'll move the variables
    'bounds': [[51, 500], # Nodes
               [25, 50],   # Joints
               [0, 1],     # Probability
               [0, 1]      # Civilian Weight
              ],
    # I don't want to group any of these variables together
    'groups': None
    }

number_of_trajectories = 100
sample = ms.sample(morris_problem, number_of_trajectories, num_levels=4, grid_jump=2)
```
