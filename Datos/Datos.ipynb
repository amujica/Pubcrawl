```python
#Recorremos cada columna (cada agente) y apuntamos su edad y genero
genders_step_0 = []

for column in genders:
    #print('Columna: {}'.format(column))
    for index, row in genders.iterrows():
        
        genders_step_0.append(row[column])
        break
 

genders_fight = [0,0]

for column in data_fights:
    #print('Agente: {}'.format(column))
    for index, row in data_fights.iterrows():
        if row[column]=='1':
            #print(genders_step_0[int(column)])
            if genders_step_0[int(column)]=='male':
                genders_fight[0]=genders_fight[0]+1
            else:
                genders_fight[1]=genders_fight[1]+1
            #Agente column, en paso index estaba en...
            #print('Para el paso {}'.format(index))
            #print(ages_fight)
            
            break
          
print(genders_fight)        
tot = 0
for index in range(len(genders_fight)):
    tot += genders_fight[index]
    
#print(tot)
for index in range(len(genders_fight)):
    genders_fight[index] = round((genders_fight[index]/tot)*100,2)
print(genders_fight)


#Recorremos cada columna (cada agente) y apuntamos su edad y genero
genders_step_0_2 = []

for column in genders_2:
    #print('Columna: {}'.format(column))
    for index, row in genders_2.iterrows():
        
        genders_step_0_2.append(row[column])
        break
 

genders_fight_2 = [0,0]

for column in data_fights_2:
    #print('Agente: {}'.format(column))
    for index, row in data_fights_2.iterrows():
        if row[column]=='1':
            #print(genders_step_0[int(column)])
            if genders_step_0_2[int(column)]=='male':
                genders_fight_2[0]=genders_fight_2[0]+1
            else:
                genders_fight_2[1]=genders_fight_2[1]+1
            #Agente column, en paso index estaba en...
            #print('Para el paso {}'.format(index))
            #print(ages_fight)
            
            break
          
print(genders_fight_2)          
tot_2 = 0
for index in range(len(genders_fight_2)):
    tot_2 += genders_fight_2[index]
    
#print(tot_2)
for index in range(len(genders_fight_2)):
    genders_fight_2[index] = round((genders_fight_2[index]/tot_2)*100,2)
print(genders_fight_2)


```
